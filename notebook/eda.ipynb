# Multivariate Analysis

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Hitung korelasi hanya untuk kolom numerik
corr = df[num_cols].corr()

# Buat mask untuk segitiga bawah
mask = np.triu(np.ones_like(corr, dtype=bool))

# Plot heatmap dengan mask
plt.figure(figsize=(12, 8))
sns.heatmap(corr, mask=mask, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Heatmap')
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

# PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(df[num_cols])

plt.scatter(pca_result[:, 0], pca_result[:, 1], c=df['Depression'], cmap='viridis')
plt.title('PCA: Depresi')
plt.show()

# t-SNE
tsne = TSNE(n_components=2, random_state=42)
tsne_result = tsne.fit_transform(df[num_cols])

plt.scatter(tsne_result[:, 0], tsne_result[:, 1], c=df['Depression'], cmap='viridis')
plt.title('t-SNE: Depresi')
plt.show()
```
